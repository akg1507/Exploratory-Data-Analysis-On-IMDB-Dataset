# 4 . Make a table of all the producers and directors of each movie. Find the top 3 producers who have produced movies with the highest average RoI?

```python
df.columns
```

```python
no_na_crew_df = df[df['crew'].notnull()]
len(no_na_crew_df) 
```

```python
def prod_dir(row) :
  try :
    producers_list = []
    directors_list = []
    for elem in row['crew'] :
      if elem['job'] == 'Producer' :
        producers_list.append(elem['name']) 
      if elem['job'] == 'Director' :
        directors_list.append(elem['name'])
     
    row['producers'] = producers_list
    row['directors'] = directors_list
    return row

  except :
    return []
```

```python
df.head(2)
```

```python
prod_dir(no_na_crew_df.iloc[1])
```

```python
df1 = no_na_crew_df.apply(prod_dir, axis = 1) 
```

```python
df1.shape
```

```python
df1[['original_title','producers','directors']].set_index('original_title')
```

```python
type(df1['producers'])
```

```python
producer_dict = {}
# iterating over each row of df 
for index, row in df1.iterrows():
# iterarting over producer column of each row
  for prod in row['producers'] :
#updating producer dictionary for each producer based on his roi of his movies
    if prod not in producer_dict.keys() :
      producer_dict[prod] = [row['roi']]
    else :
      value_of_prod_key = producer_dict[prod]
      value_of_prod_key.append(row['roi'])
      producer_dict[prod] = value_of_prod_key
```

```python
top_3_producers = pd.DataFrame.from_dict(producer_dict, orient = 'index').mean(axis = 1).reset_index().rename({'index':'PRODUCER', 0:'avg(ROI)'},  axis = 1).sort_values(by = 'avg(ROI)',ascending = False).iloc[0:3]
```

```python
top_3_producers
```

# 5. Which actor has acted in the most number of movies? Deep dive into the movies, genres and profits corresponding to this actor.

```python
no_na_cast_df = df[df['cast'].notnull()]
```

```python
actor_movie_dict = {}
for index , row in  no_na_cast_df.iterrows() :
  #print(row)
  for elem in row['cast'] :
    if elem['name'] not in actor_movie_dict.keys() :
      actor_movie_dict[elem['name']] = [row['imdb_id']]
    else :
      value_of_actor_key = actor_movie_dict[elem['name']]
      value_of_actor_key.append((row['imdb_id']))
      actor_movie_dict[elem['name']] = value_of_actor_key
```

```python
actor_movies_num = pd.DataFrame.from_dict(actor_movie_dict , orient = 'index').count(axis = 1).reset_index().rename(columns = {'index':'Actor',0 :'number of movies'}).sort_values(by = 'number of movies', ascending = False)
```

```python
top_actors = actor_movies_num.loc[actor_movies_num['number of movies'] == actor_movies_num['number of movies'].max(),['Actor','number of movies']]
top_actors
```

```python
df_dict = {'actor': [] , 'movies' : [], 'profits' : [], 'roi' : [], 'genres' : []}
for i, r in no_na_cast_df.iterrows() :
  for e in r['cast'] :
    if e['name'] in list(top_actors['Actor']) :
      df_dict['actor'].append(e['name'])
      df_dict['movies'].append(r['original_title'])
      df_dict['profits'].append(r['profit'])
      df_dict['roi'].append(r['roi'])
      genres_set = set()
      for f in r['genres'] :
        genres_set.add(f['name'])
      df_dict['genres'].append(genres_set)
```

```python
top_actors_info = pd.DataFrame(df_dict)
```

# 6. Top 3 directors prefer which actors the most?


```python
df.head(2)
```

```python
df.info()
```

```python
no_na_crew_df = df[df['crew'].notnull()]
no_na_crew_df.head()
```

```python
def dir(row) :
  try :
    directors_list = []
    for elem in   row['crew'] :
      if elem['job'] == 'Director' :
        directors_list.append(elem['name'])
    row['directors'] = directors_list
    return row

  except :
    return []
```

```python
dir(no_na_crew_df.iloc[0])
```

```python
no_na_crew_df = no_na_crew_df.apply(dir , axis = 1) 
```

```python
dir_info = {}
for i, r in no_na_crew_df.iterrows() :
  for d in r['directors'] :
    if d not in dir_info.keys() :
      dir_info[d] = [r['roi']]
    else :
      v = dir_info[d]
      v.append(r['roi'])
      dir_info[d] = v
```

```python
df['roi'].mean()
```

```python
directors_roi = pd.DataFrame.from_dict(dir_info, orient = 'index').mean(axis = 1).reset_index()
```

```python
sorted_drcts_roi = directors_roi.rename(columns = {'index':'name of directors',0:'avg roi'}).sort_values(by = 'avg roi',ascending = False)
```

```python
top_3_directors = sorted_drcts_roi.iloc[0:3]
top_3_directors
```
